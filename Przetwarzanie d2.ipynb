#F_CENA

```python
import pandas as pd
import numpy as np
```

```python
F_CENA = pd.read_csv('/content/F_CENA.csv',
                     encoding='utf-8',
                     sep='&')
F_CENA
```

```python
remove_col_list = ['Unnamed: 0.2',
                   'Unnamed: 0.1',
                   'Unnamed: 0',
                   'DATA_IMP']
```

```python
F_CENA = F_CENA.drop(remove_col_list,
                     axis=1)
F_CENA
```

```python
F_CENA.drop(['PRED_ID'],
            axis=1,
            inplace=True)
```

```python
F_CENA
```

```python
F_CENA.info()
```

```python
F_CENA['DZIEN_DATA'] = pd.to_datetime(F_CENA['DZIEN_DATA'],
                                      format ='%Y-%m-%d')
F_CENA
```

```python
F_CENA.info()
```

```python
F_CENA_filter_nan =  F_CENA[F_CENA.isnull().any(axis=1)]
F_CENA_filter_nan
```

```python
col_with_null_list =  F_CENA.columns[F_CENA.isna().any()].tolist()
col_with_null_list
```

```python
for i in np.arange(0, len(col_with_null_list)):
  F_CENA[col_with_null_list[i]] =  F_CENA[col_with_null_list[i]].fillna(method='ffill')
```

```python
cols = [col for col in F_CENA.columns if '_ID' in col]
cols
```

```python
F_CENA = F_CENA.set_index(cols)
F_CENA
```

```python
F_CENA_groupby_TOW_ID = F_CENA[F_CENA['CENS_CEN_SYS'] > 0].groupby(['SKL_ID','TOW_ID'])['CENS_CEN_SYS'].agg(['max', 'min','mean'])
F_CENA_groupby_TOW_ID
```

```python
F_CENA.reset_index(drop=False, inplace=True)
```

```python
F_CENA_groupby_TOW_ID = F_CENA[(F_CENA['CENS_CEN_SYS'] > 0) & (F_CENA['SKL_ID'] == 1) & (F_CENA['DZIEN_DATA'] >= '2021-01-01')].groupby(['SKL_ID', 'TOW_ID'])['CENS_CEN_SYS'].agg(['max',
                                                                        'min',
                                                                        'mean'])
F_CENA_groupby_TOW_ID
```

```python
F_CENA
```

```python
## Sprawdzenie statystyk dla cen wyższych niż zero
F_CENA[(F_CENA['CENS_CEN_SYS'] > 0)].groupby(['SKL_ID', 'TOW_ID', F_CENA['DZIEN_DATA'].dt.year])['CENS_CEN_SYS'].agg(['max', 'min', 'mean']).reset_index()

```

#F_OTW

```python
F_OTW = pd.read_parquet('/content/F_OTW.parquet')
F_OTW
```

```python
F_OTW[F_OTW.duplicated(subset=['SKL_ID', 'DZIEN_ID'],
                       keep='first')]
```

```python
F_OTW.drop_duplicates(inplace=True)
```

```python
F_OTW['OS_OTW_NAZ'] = pd.to_datetime(F_OTW['OS_OTW_NAZ'],
                                     errors='coerce',
                                     format='%H:%M:%S'
                                     )


F_OTW['OS_ZAMK_NAZ'] = pd.to_datetime(F_OTW['OS_ZAMK_NAZ'],
                                     errors='coerce',
                                     format='%H:%M:%S'
                                     )

F_OTW
```

```python
F_OTW.info()
```

```python
F_OTW['OS_ROZN_H'] = (F_OTW['OS_ZAMK_NAZ'] - F_OTW['OS_OTW_NAZ']).dt.total_seconds() / 3600
F_OTW['OS_ROZN_H']
```

```python
F_OTW['OS_ROZN_H'].fillna(0, inplace=True)
F_OTW
```
